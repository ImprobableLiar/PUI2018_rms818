```python
tr = pd.read_csv("%s/tapresult.csv"%os.getenv("PUIDATA"))
```

```python
tr.head(7)
```

```python
tr_2c = tr[['Lead First Draw (mg/L)', 'Copper First Draw (mg/L)']]
tr_2c.columns = ['Lead', 'Copper']
tr_2c
```

```python
tr_2c.Lead = pd.to_numeric(tr_2c.Lead, errors='coerce')
tr_2c.Copper = pd.to_numeric(tr_2c.Copper, errors='coerce')
tr_2c.dtypes
```

```python
tr_2c.plot.scatter(x = "Copper", y = "Lead", title = "Lead and Copper At-The-Tap Results")
ylabel('Lead First Draw (mg/L)')
xlabel('Copper First Draw (mg/L)')
```

### Figure 1.
Scatter plot of lead and copper levels in the first draw of the tap at different locations. 

It should be noted that there are locations with unusually high levels of copper and/or lead in their tap water.

```python
tr_3c = tr[['Date Collected','Lead First Draw (mg/L)', 'Copper First Draw (mg/L)']]
tr_3c.columns = ['Date', 'Lead', 'Copper']
tr_3c
```

```python
tr_3c.Date = pd.to_datetime(tr_3c.Date, errors='coerce')
tr_3c.Lead = pd.to_numeric(tr_3c.Lead, errors='coerce')
tr_3c.Copper = pd.to_numeric(tr_3c.Copper, errors='coerce')
tr_3c.dtypes
```

```python
ax = tr_3c.plot(x='Date', y='Lead');
ax.set_ylabel("Lead First Draw (mg/L)");
ax.set_title("At-The-Tap Lead Levels over Time");
```

### Figure 2.
Graph of lead levels in the first draw of the tap over time.

There were spikes in lead levels in 2015 but lead levels seem to have reduced afterwards.

```python
ax = tr_3c.plot(x='Date', y='Copper');
ax.set_ylabel("Copper First Draw (mg/L)");
ax.set_title("At-The-Tap Copper Levels over Time");
```

### Figure 3.
Graph of copper levels in the first draw of the tap over time.

Copper in water seems to be prevalent only after 2015, with occasional spikes from 2016 onwards.
