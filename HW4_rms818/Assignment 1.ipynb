## Normal Distribution

```python
mymean = 1000
```

```python
np.random.seed(303)
mysize = np.random.randint(11, 2000, 100)

md = {}

md['normaldist'] = {}

for n in mysize:
    md['normaldist'][n] = np.random.normal(loc = mymean, scale = 1.0, size = n)
    
md['normaldist']['means'] = {}
axnormaldist_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['normaldist']:
    if not type(nn) == str:
        md['normaldist']['means'][nn] = md['normaldist'][nn].mean()
        axnormaldist_mu_n.plot(nn, md['normaldist']['means'][nn], 'o')
        axnormaldist_mu_n.set_xlabel('sample size', fontsize=18)
        axnormaldist_mu_n.set_ylabel('sample mean', fontsize=18)
        axnormaldist_mu_n.set_title('Normal', fontsize=18)
        axnormaldist_mu_n.plot([min(mysize), max(mysize)], [1000, 1000], 'k')
```

## Poisson Distribution

```python
np.random.seed(303)
mysize = np.random.randint(11, 2000, 100)

md['poissondist'] = {}

for n in mysize:
    md['poissondist'][n] = np.random.poisson(lam = mymean, size = n)
    
md['poissondist']['means'] = {}
axpoissondist_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['poissondist']:
    if not type(nn) == str:
        md['poissondist']['means'][nn] = md['poissondist'][nn].mean()
        axpoissondist_mu_n.plot(nn, md['poissondist']['means'][nn], 'o')
        axpoissondist_mu_n.set_xlabel('sample size', fontsize=18)
        axpoissondist_mu_n.set_ylabel('sample mean', fontsize=18)
        axpoissondist_mu_n.set_title('Poisson', fontsize=18)
        axpoissondist_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

## Binomial Distribution

```python
np.random.seed(303)
mysize = np.random.randint(11, 2000, 100)

md['binomialdist'] = {}

for n in mysize:
    md['binomialdist'][n] = np.random.binomial(n = 2000, p = 0.5, size = n)
    
md['binomialdist']['means'] = {}
axbinomialdist_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['binomialdist']:
    if not type(nn) == str:
        md['binomialdist']['means'][nn] = md['binomialdist'][nn].mean()
        axbinomialdist_mu_n.plot(nn, md['binomialdist']['means'][nn], 'o')
        axbinomialdist_mu_n.set_xlabel('sample size', fontsize=18)
        axbinomialdist_mu_n.set_ylabel('sample mean', fontsize=18)
        axbinomialdist_mu_n.set_title('Binomial', fontsize=18)
        axbinomialdist_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

## Chi-square Distribution

```python
np.random.seed(303)
mysize = np.random.randint(11, 2000, 100)

md['chisqdist'] = {}

for n in mysize:
    md['chisqdist'][n] = np.random.chisquare(df = mymean, size = n)
    
md['chisqdist']['means'] = {}
axchisqdist_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisqdist']:
    if not type(nn) == str:
        md['chisqdist']['means'][nn] = md['chisqdist'][nn].mean()
        axchisqdist_mu_n.plot(nn, md['chisqdist']['means'][nn], 'o')
        axchisqdist_mu_n.set_xlabel('sample size', fontsize=18)
        axchisqdist_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisqdist_mu_n.set_title('Chi-Square', fontsize=18)
        axchisqdist_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

```python
# underlying normal distribution mean set to 6.40775/4.10517 and sigma set to 1.0 so that lognormal sample mean will be 1000

np.random.seed(303)
mysize = np.random.randint(11, 2000, 100)

md['lognormaldist'] = {}

for n in mysize:
    md['lognormaldist'][n] = np.random.lognormal(mean = 6.40775, sigma = 1.0, size = n)
    
md['lognormaldist']['means'] = {}
axlognormaldist_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['lognormaldist']:
    if not type(nn) == str:
        md['lognormaldist']['means'][nn] = md['lognormaldist'][nn].mean()
        axlognormaldist_mu_n.plot(nn, md['lognormaldist']['means'][nn], 'o')
        axlognormaldist_mu_n.set_xlabel('sample size', fontsize=18)
        axlognormaldist_mu_n.set_ylabel('sample mean', fontsize=18)
        axlognormaldist_mu_n.set_title('Log-Normal', fontsize=18)
        axlognormaldist_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```
